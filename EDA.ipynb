```python
import seaborn as sns
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df[["Age","Fare"]],palette="dark")
plt.title("boxplot for outlier detection")
plt.show()
```

```python
plt.figure(figsize=(25,10))
sns.countplot(x="Survived",data=df,palette="pastel")
plt.title("survival count",fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(30,15))
sns.histplot(df["Age"],bins=30,kde=True,color="red")
plt.title("Age distribution",fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(40,25))
sns.scatterplot(data=df[["Age","Fare"]],palette="muted")
plt.title("scatterplot for outlier detection")
plt.show()
```

```python
plt.figure(figsize=(15,3))
sns.violinplot(x="Survived",y="Fare",data=df,palette="dark")
plt.title("violin plot:fare distribution by survival",fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.kdeplot(df["Age"],shade=True,color="yellow")
plt.title("density plot:Age Distribution",fontsize=14)
plt.show()
```
