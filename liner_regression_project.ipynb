```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data=pd.read_csv("/content/Salary_Data.csv")
data.isnull().sum()
data.head(8)
data.describe()
```

```python
x=data["YearsExperience"]
y=data["Salary"]
```

```python
plt.figure(figsize=(20,18))
sns.boxplot(data=data[["YearsExperience","Salary"]])
plt.title("boxsize for outlier detection")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x="YearsExperience",data=data,palette="pastel")
plt.title("YearsExperience vs Salary",fontsize=13)
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.histplot(data["YearsExperience"],bins=30,kde=True,color="blue")
plt.title("YearsExperience vs Salary",fontsize=13)
plt.show()
```

```python
plt.figure(figsize=(9,3))
sns.scatterplot(x="YearsExperience",y="Salary",data=data,color="red")
plt.title("YearsExperience vs Salary",fontsize=14)
plt.show()
```
